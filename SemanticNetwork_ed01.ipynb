```python
from nltk.corpus import wordnet as wn
import networkx as nx
```

```python
words = ['green','red','blue','dog','wolf','cat','tiger','lion']
words_hyp = []
for i in words:
    words_hyp.append((i,(wn.synsets(i)[0].hypernyms()[0].name().split(".")[0])))
```

```python
words_hyp
```

```python
plt.rcParams["figure.figsize"] = (20,10)
import networkx as nx
import matplotlib.pyplot as plt

#This list shows the nodes connectios: 0 with 1, 1 with 5 and with 2...
points_list = words_hyp

G=nx.Graph()
G.add_edges_from(points_list)


pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G,pos,node_color="yellow",edgecolors="black")
nx.draw_networkx_edges(G,pos,edge_color="black")
nx.draw_networkx_labels(G,pos,font_color="red")

plt.axis("off")
plt.show()
```

```python
import graphviz
```

```python
# All Hypernyms of several synsets with its branches

def semantic_tree (s_list):
    dot = graphviz.Digraph(strict=True)
    for i in s_list:
        item = i + '.n.01'
        chain = [i]
        an = wn.synset(item)

        hyp = an.hypernyms()
        while hyp != []:
            chain.append(hyp[0].name().split(".")[0])
            hyp = hyp[0].hypernyms()

        for i in chain:
            dot.node(i)

        for i in range(len(chain)-1):
            dot.edge(chain[i], chain[i+1])

    return(dot)
```

```python
semantic_tree(['dog', 'cat', 'human', 'fish', 'apple', 'stone', 'pelican'])
```

```python
# All Hypernyms of several synsets with its branches

def td_semantic_tree (s_list):
    dot = graphviz.Digraph(strict=True)
    for i in s_list:
        item = i + '.n.01'
        chain = [i]
        an = wn.synset(item)

        hyp = an.hypernyms()
        while hyp != []:
            chain.append(hyp[0].name().split(".")[0])
            hyp = hyp[0].hypernyms()

        chain = list(reversed(chain)) # List reversed to make the tree top-down
        
        for i in chain:
            dot.node(i)

        for i in range(len(chain)-1):
            dot.edge(chain[i], chain[i+1])

    return(dot)
```

```python
td_semantic_tree(['dog', 'cat', 'human', 'fish', 'apple', 'iron', 'pelican', 'orange', 'lemon'])
```

```python
# All Hypernyms of several synsets with its branches

def lr_semantic_tree (s_list):
    dot = graphviz.Digraph(strict=True)
    dot.attr(rankdir='LR')
    for i in s_list:
        item = i + '.n.01'
        chain = [i]
        an = wn.synset(item)

        hyp = an.hypernyms()
        while hyp != []:
            chain.append(hyp[0].name().split(".")[0])
            hyp = hyp[0].hypernyms()

        for i in list(reversed(chain)):
            dot.node(i)

        for i in range(len(chain)-1):
            dot.edge(list(reversed(chain))[i], list(reversed(chain))[i+1])

    return(dot)
```

```python
lr_semantic_tree(['dog', 'cat', 'human', 'fish', 'apple', 'iron', 'pelican'])
```

```python
# Other engines

def engine_semantic_tree (s_list):
    dot = graphviz.Digraph(strict=True, engine='neato')
    for i in s_list:
        item = i + '.n.01'
        chain = [i]
        an = wn.synset(item)

        hyp = an.hypernyms()
        while hyp != []:
            chain.append(hyp[0].name().split(".")[0])
            hyp = hyp[0].hypernyms()

        for i in list(reversed(chain)):
            dot.node(i)

        for i in range(len(chain)-1):
            dot.edge(list(reversed(chain))[i], list(reversed(chain))[i+1])


    return(dot)
```

```python
engine_semantic_tree(['dog', 'cat', 'human', 'fish', 'apple', 'iron', 'pelican'])
```
